 # Split data into X & Y

```python
X = ns.random.rand(4649,5)
Y = ns.random.rand(4649)
```

```python
# Train_Test_Split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Standardize the Data
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
# Support Vector Regressor
SV1 = SVR()
SV1.fit(X_train,Y_train)
```

```python
# Ada Boost Regressor
Adc1 = AdaBoostRegressor(n_estimators=50,learning_rate=1,random_state=42)
Model_1 = Adc1.fit(X_train,Y_train)
```

```python
# KNN Regressor
model1 = KNeighborsRegressor()
model1.fit(X_train,Y_train)
```

```python
# Decision Tree Regressor
DT_Regs = DecisionTreeRegressor(random_state=42)
DT_Regs.fit(X_train,Y_train)
```

```python
# K Means
k_means = KMeans()
k_means.fit(X_train,Y_train)
```

# Predictions from our model

```python
Predictions = DT_Regs.predict(X_test)
Predictions
```

```python
# Evaluate the Model
mse = mean_squared_error(Y_test,Predictions)
print("Mean Squared Error:",mse)
```
