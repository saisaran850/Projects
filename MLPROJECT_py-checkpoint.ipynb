 # Split the data into X & Y

```python
X = nd.delete(df,0,axis=0)
Y = nd.delete(df,0,axis=0)
```

```python
# Train_Test_Split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Standardize the Data
Scaler = StandardScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

 # Model Selection and Training

```python
RF_Regs = RandomForestRegressor(random_state=42)
RF_Regs.fit(X_train,Y_train)
```

```python
Lin_Regr = LinearRegression()
Lin_Regr.fit(X_train,Y_train)
```

```python
DT_Regs = DecisionTreeRegressor(random_state=42)
DT_Regs.fit(X_train,Y_train)
```

# Model Evaluation

```python
def evaluate_model(model,X_test,Y_test):
    Y_pred = model.predict(X_test)
    rmse = nd.sqrt(mean_squared_error(Y_test,Y_pred))
    r2 = r2_score(Y_test,Y_pred)
    return rmse,r2
# Evaluate Models
Lin_rmse,Lin_r2 = evaluate_model(Lin_Regr,X_test,Y_test)
DT_rmse,DT_r2 = evaluate_model(DT_Regs,X_test,Y_test)
RF_rmse,RF_r2 = evaluate_model(RF_Regs,X_test,Y_test)
print(f"Linear Regression RMSE: {Lin_rmse},R2: {Lin_r2}")
print(f"Decision Tree RMSE: {DT_rmse},R2: {DT_r2}")
print(f"Random Forest RMSE: {RF_rmse},R2: {RF_r2}")
```

# Hyper Parameter Tuning

```python
param_grid = {
    'n_estimators' : [100,200,300],'max_depth' : [None,10,20,30],'min_samples_split' : [2,5,10]
}
grid_search = GridSearchCV(estimator=RF_Regs,param_grid=param_grid,cv=3,n_jobs=-1,verbose=2)
grid_search.fit(X_train,Y_train)
best_rf = grid_search.best_estimator_
# Evaluate the Model
best_rf_rmse,best_rf_r2 = evaluate_model(best_rf,X_test,Y_test)
print(f"Best Random Forest RMSE: {best_rf_rmse},R2: {best_rf_r2}")
```
