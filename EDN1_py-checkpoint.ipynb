```python
import matplotlib.pyplot as plt
labels = ["Maharashtra","Hyderabad","Tamilnadu"]
sizes = [500,600,700]
plt.pie(sizes,labels=labels,autopct = "%1.1f%%",startangle = 95)
plt.title("PIE CHART")
plt.show()
```

```python
import matplotlib.pyplot as plt
D = [111,145,167,178,189]
J1,J2 = [201,225,245,265,280],[315,367,378,395,401]
plt.fill_between(D,J1,J2,color = "Yellow",alpha = 0.6,label = "Area 1-2")
plt.plot(D,J1,label = "Line1",marker = "o")
plt.plot(D,J2,label = "Line2",marker = "o")
plt.xlabel("D-axis"),plt.ylabel("J-axis"),plt.title("Area Plot")
plt.legend(),plt.show()
```
